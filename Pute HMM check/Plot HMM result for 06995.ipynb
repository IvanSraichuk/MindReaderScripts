```python
import numpy as np
from IPython.display import display
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import neurokit2 as nk
import warnings

%matplotlib inline
```

```python
traceback1 = '06995ECG.txt50P_ECG1_traceback.csv'
traceback2 = '06995ECG.txt25P_ECG1_traceback.csv'
signal_path = '06995ECG.txt'
```

```python
traceback1 = np.loadtxt(traceback1, dtype=int)
traceback2 = np.loadtxt(traceback2, dtype=int)
arr = np.loadtxt(signal_path, dtype=float)
```

```python
np.unique(traceback1, return_counts=True)
```

```python
np.unique(traceback2, return_counts=True)
```

```python
colors = ["#0000FF", "#00FF00", "#FF0066", "#FFFF00", "#FF00FF", '#99CCFF' ]
for i in range(4000, 20000, 2000):
    traceback = traceback1[i:i+2000]
    points=[]
    signal = arr[i:i+2000, 0]
    for j in range(len(traceback)):
        if traceback[j] > 1:
            points.append([j, signal[j], colors[traceback[j]]])
    if len(points) > 0:
        plt.figure(figsize=(20, 3))
        plt.plot(signal)
        for i in range(len(points)):
            plt.scatter(points[i][0], points[i][1], color=points[i][2])    
        plt.show()
```

```python
for i in range(4000, 20000, 2000):
    traceback = traceback2[i:i+2000]
    points=[]
    signal = arr[i:i+2000, 0]
    for j in range(len(traceback)):
        if traceback[j] > 1:
            points.append([j, signal[j], colors[traceback[j]]])
    if len(points) > 0:
        plt.figure(figsize=(20, 4))
        plt.plot(signal)
        for i in range(len(points)):
            plt.scatter(points[i][0], points[i][1], color=points[i][2])    
        plt.show()
```
