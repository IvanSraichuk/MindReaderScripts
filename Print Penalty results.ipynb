````python
import numpy as np
from IPython.display import display
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import neurokit2 as nk
import warnings
import io
%matplotlib inline
````

````python
colors = {
    'Accuracy':'blue',
    'Specificity':'black',
    'AF_Accuracy':'green',
    'MCC':'red',
    'FDR':'yellow', 
    'Sensitivity':'Pink', 
    'FOR':'Orange', 
    'Penalty':'Purple', 
    'NPV':'Brown', 
    'FPR':'DarkKhaki', 
    'FNR':'MediumPurple', 
    'FScore':'Olive',
    'Accuracy2': 'Thistle'
}
def PlotMeasures(values):
    plt.figure(figsize=(20, 6.25))
    plt.xlim(0,1300)
    added_labels = [] # in order to avoid duplicated in a legend
    for d in values:
        for key, value in d.items():
            if not value or key == 'Penalty':
                continue
            if key in added_labels:
                plt.scatter(d['Penalty'], value, color = colors[key])
            else:
                plt.scatter(d['Penalty'], value, color = colors[key], label = key)
                added_labels.append(key)
    plt.legend(loc="upper right")
    plt.show()
````
